```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_start = [1,2,3,4,6,7,8]
x = np.array(x_start)
y = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Exponential Graph')

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_xlabel('Y')
axes2.set_ylabel('X')
axes2.set_title('Exponential Graph 2')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=[8,2])

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.savefig('2_figs.png',dpi=200)
```

```python
fig = plt.figure()

axis = fig.add_axes([0.2,0.2,0.6,0.6])

axis.plot(x,y,label='Exponential')
axis.plot(y,x,label='Reverse Exponential')

axis.legend(loc=0)
```
